Using Statistical Methods:

Z-Score

Interquartile Range (IQR)

```python
import numpy as np
import pandas as pd

# Sample Data
data = pd.DataFrame({'Value': [10, 12, 14, 18, 21, 22, 24, 100]})

# Z-Score Method
from scipy.stats import zscore
data['Z-Score'] = zscore(data['Value'])
print(data)

# IQR Method
Q1 = data['Value'].quantile(0.25)
Q3 = data['Value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(data[(data['Value'] < lower_bound) | (data['Value'] > upper_bound)])
```

Using Visualization:

Boxplots

Scatter plots

```python
import matplotlib.pyplot as plt

# Boxplot
plt.boxplot(data['Value'])
plt.title("Boxplot to Detect Outliers")
plt.show()
```

```python


# Scatter Plot
plt.scatter(range(len(data)), data['Value'])
plt.title("Scatter Plot to Detect Outliers")
plt.show()
```

Removal of Outliers:

Using filtering techniques.

```python
# Removing Outliers
filtered_data = data[(data['Value'] >= lower_bound) & (data['Value'] <= upper_bound)]
print(filtered_data)
```

Imputation Techniques:

Replacing with mean/median.

```python
# Replace outliers with median
median_value = data['Value'].median()
data['Value'] = np.where(
    (data['Value'] < lower_bound) | (data['Value'] > upper_bound),
    median_value,
    data['Value']
)
print(data)
```

Capping/Flooring:

```python
# Capping outliers
data['Value'] = np.clip(data['Value'], lower_bound, upper_bound)
print(data)
```

Transformation Techniques:

Log Transformation

Box-Cox Transformation

Log Transformation



```python
data['Log_Transformed'] = np.log(data['Value'])
print(data)
```
